```python
# Performance Evaluation Programming
# Task: kNN classifier (k=5) on Iris dataset + evaluation metrics

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.multiclass import OneVsRestClassifier

# 1. Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Standardize features for better performance of kNN
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Binarize labels for ROC curve (one-vs-rest)
y_bin = label_binarize(y, classes=[0, 1, 2])
n_classes = y_bin.shape[1]

# Train-test split
X_train, X_test, y_train, y_test, y_bin_train, y_bin_test = train_test_split(
    X, y, y_bin, test_size=0.3, random_state=42, stratify=y
)

# 2. Train kNN classifier (k=5)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# 3. Accuracy, Precision, Recall, F1
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# 4. ROC Curve and AUC (One-vs-Rest)
classifier = OneVsRestClassifier(KNeighborsClassifier(n_neighbors=5))
classifier.fit(X_train, y_bin_train)
y_score = classifier.predict_proba(X_test)

# Compute ROC curve and AUC for each class
fpr, tpr, roc_auc = dict(), dict(), dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_bin_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
plt.figure(figsize=(7, 6))
colors = ['blue', 'green', 'red']
for i, color in zip(range(n_classes), colors):
    plt.plot(
        fpr[i], tpr[i], color=color, lw=2,
        label=f'ROC curve of class {iris.target_names[i]} (AUC = {roc_auc[i]:.2f})'
    )

plt.plot([0, 1], [0, 1], 'k--', lw=2)  # diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for Iris (kNN, k=5)")
plt.legend(loc="lower right")
plt.show()
```
